```python
# funtion to convert date int -> datetime -> float (just year yyyy)
def year_dtype_to_datetime(dataframe:pd.DataFrame, columns:list, format:str="%Y") -> pd.DataFrame:
    """
    Input:
        dataframe: Input pandas dataframe
        columns: list of columns in dataframe to be converted
        format: datetime pattern to get year
    Output:
        dataframe: dataframe with dtypes from date columns converted to float years yyyy format 
    """
    dataframe[columns] = dataframe[columns].apply(pd.to_datetime, format=format, errors="coerce")
    dataframe[columns] = dataframe[columns].apply(lambda x: x.dt.year)
    return dataframe
```

```python
# function to replace empty strings with NaN
def replace_empty_strings(dataframe:pd.DataFrame, columns:list) -> pd.DataFrame:
    """
    Input:
        dataframe: Input pandas dataframe
        columns: list of columns in dataframe to be changed
    Output:
        dataframe: dataframe with empty strings data from columns replace with NaN null values
    """
    dataframe[columns] = dataframe[columns].replace("", np.NaN, regex=False)
    return dataframe
```

```python
#function to change dtype to datetime with format yyyy-mm-dd
def date_str_to_datetime(dataframe:pd.DataFrame, column:str) -> None:
    """
    Input:
        dataframe: Input pandas dataframe
        column: column in dataframe to be changed
    Action:
        dataframe with column changed to datatype datetime with format %Y-%m-%d
    """
    dataframe[column] = pd.to_datetime(dataframe[column].astype(str), format="%Y-%m-%d")
```

```python
# function to count NaN values in a dataframe
def look_null_values(dataframe:pd.DataFrame, column:str) -> int:
    """
    Counts the NaN ot NaT values in the dataframe column
    """
    print(f"Looking for NaN ot NaT values in {column} column")
    null_count = 0
    value_array = dataframe[column]
    for value in value_array:
        if pd.isnull(value) or value == pd.NaT:
            null_count += 1
    return null_count
```

```python
# function to summarise the datatype and NaN values from a dataframe
def df_nan_summary(dataframe:pd.DataFrame) -> None:
    columns = dataframe.keys()
    for column in columns:
        null_count = look_null_values(dataframe, column)
        column_len = len(dataframe[column])
        print("Column datatype: " + str(type(dataframe[column][0])))
        print("Total NaN values: " + str(null_count))
        print("Total number of values: " + str(column_len) + "\n")
```

# Merging Dataframes

**T6: Merging the 2 dataframes using 'ConstituentID' as key.**

```python
# Task 6: merge two dataframes using ConstituentID as the key
```

In order to merge artworks dataset with artists datset, artworks with multiple artists need to be seperated into individual rows with their respective ConstituentID

```python
fixed_cols = list(df_artworks.keys())
fixed_cols.remove('ConstituentID')

df_artworks_normalised = df_artworks.set_index(fixed_cols).apply(lambda x: x.str.split(",").explode()).reset_index()
print(df_artworks.shape, df_artworks_normalised.shape)
```

```python
df_artists.columns
```

```python
df_artworks_normalised.columns
```

To merge artworks with artists --> need to drop intersecting columns

```python
# removing intersecting columns from artworks df
INTERSECTING_COLS = ['ArtistBio', 'Nationality', 'BeginDate', 'EndDate', 'Gender']
df_artworks_normalised = df_artworks_normalised.drop(columns=INTERSECTING_COLS)
```

```python
df_artworks_normalised.shape
```

changing ConstituentId to numeric datatype to be merged

```python
type(df_artworks_normalised["ConstituentID"][0])
```

```python
df_artworks_normalised["ConstituentID"] = pd.to_numeric(df_artworks_normalised["ConstituentID"])
```

merging both dataframes

```python
df_join = pd.merge(df_artworks_normalised, df_artists, on="ConstituentID", how="left")
df_join.shape
```
